```python
#QUESTION TWO: 
#GITHUB API:

#CLIENT ID: 4c1717eba29036598fe7
#CLIENT SECRET: a802db87808b7aa4a66f7d662c78656406e1c80a
# github.py

import requests #imports request module so we can use the requests.get method and other methods

# REPLACE the following variables with your Client ID and Client Secret
CLIENT_ID = "4c1717eba29036598fe7" #sets client id variable found via github APP
CLIENT_SECRET = "a802db87808b7aa4a66f7d662c78656406e1c80a" #sets client secret variable found via github APP

# REPLACE the following variable with what you added in the
# "Authorization callback URL" field
REDIRECT_URI = "https://httpbin.org/anything" #authorization link, sort of like a middle ground to let information 
#go back and fourth between API and code to grab info

def create_oauth_link(): #creates function that creates the oauthentication link
    params = { #starts dictioniary for params variable
        "client_id": CLIENT_ID, #sets client id as stated above in dictionary format
        "redirect_uri": REDIRECT_URI, #sets redirect_uri as stated above in dictionary format
        "scope": "user", #sets scope as user in dictionary format
        "response_type": "code", #sets response to code in dictionary format, which for from lines 21-24 is smart because
        #the person coding can get specific key values just by stating its names instead of indexes which can be 
        #ambiguous
    } #closes dictionary

    endpoint = "https://github.com/login/oauth/authorize" #makes the location where the API gets the info, is the 
    #"point of contact between API client and API server" (work cited below)
    response = requests.get(endpoint, params=params) #gets the response of the oauth link based on the parameters
    #written above, requests.get gets the info 
    return response.url #returns the response with .url, which gets the endpoint info and lets the computer know 
    #what you want to receive/interact with

def exchange_code_for_access_token(code=None): #makes function to trade access token with access code
    params = { #sets parameter dictionary
        "client_id": CLIENT_ID, #makes key value pair for client ID
        "client_secret": CLIENT_SECRET, #makes key value pair for client secret
        "redirect_uri": REDIRECT_URI, #makes key value pair for redirect_uri
        "code": code, #makes key value pair for code which we will get later on when following link
    }

    headers = {"Accept": "application/json"} #header lets info move between user and server
    endpoint = "https://github.com/login/oauth/access_token" #makes the location where the API gets the info, is the
    #'point of contact between API client and API server' (again, work cited below for endpoint)
    response = requests.post(endpoint, params=params, headers=headers).json() #requests.post used to send data to server
    #used to send code info and other parameters to retrieve the access token
    return response["access_token"] #returns the access token

def print_user_info(access_token=None): #makes function that returns user info
    headers = {"Authorization": f"token {access_token}"} #lets info move between user and server, inputs acess token to
    #get access to this information and probably for the specific user
    endpoint = "https://api.github.com/user" #again, makes location where the API gets info, makes sense that its the
    #github website because that's where we want to get the user information from
    response = requests.get(endpoint, headers=headers).json() #converts response to json and 'response' is the variable
    #we use to actually retrieve information about the user, as seen below with specific key values
    name = response["name"] #gets the name of the user by using specific key "name"
    username = response["login"] #gets the username of the user by using specific key "login"
    private_repos_count = response["total_private_repos"] #gets amount of private repos via specific key value
    print( #prints the information the server retrieved via the response json
        f"{name} ({username}) | private repositories: {private_repos_count}" #prints the information the server
        #retrieved
    ) #closes print statement
    
link = create_oauth_link() #creates variable for oauth link via oauth link function made above
print(f"Follow the link to start the authentication with GitHub: {link}") #prints statement to get code to trade for
#the access token
code = input("GitHub code: ") #makes input where you put the code found via the oauth link and print statement above
access_token = exchange_code_for_access_token(code) #creates access token via inputting code into function that gets
#the access code
print(f"Exchanged code {code} with access token: {access_token}") #prints statement saying your code was exchanged
#with the access token, and then prints the access token 
print_user_info(access_token=access_token) #prints the user information and uses access token as a parameter that is 
#automatically inputted (not needed to do manually, already done)
    
#WORK CITED: https://www.bing.com/search?pglt=673&q=what+is+an+endpoint+in+api&cvid=e8776b8421084433bff6917ed69b80ca&gs_lcrp=EgZjaHJvbWUyBggAEEUYOTIGCAEQABhAMgYIAhAAGEAyBggDEAAYQDIGCAQQABhAMgYIBRAAGEAyBggGEAAYQDIGCAcQABhAMgYICBAAGEDSAQg2NDg5ajBqMagCALACAA&FORM=ANSPA1&PC=HCTS
#link above was used to understand what endpoints are for APIs on line 27/28 and wherever endpoint = was used
```
